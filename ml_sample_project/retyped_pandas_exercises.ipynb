```python
# Import the car sales with missing values csv
car_sales_missing = pd.read_csv('car-sales-missing-data.csv')
car_sales_missing
```

```python
# Fill the missing values in the Odometer column with the mean of the column
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
car_sales_missing
```

```python
# Remove the rest of the missing data in place
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
# Create a Seats column where every row has the value of 5
car_sales_missing["Seats"] = 5
car_sales_missing
```

```python
# Create a  new column called "Engine Size" with random values between 1.3 and 4.5
seats = pd.Series([1.32, 4.4, 3.4, 3.6, 4.2, 2.3])
car_sales_missing["Seats"] = seats
car_sales_missing
```

```python
# Create a new column for the price of a car per km
car_sales_missing["Price (USD)"] = car_sales_missing["Price"].str.replace("[$,]", '', regex=True).astype(int)
car_sales_missing["Price / KM"] = car_sales_missing["Price (USD)"] / car_sales_missing["Odometer"]
car_sales_missing
```

```python
# Drop the Seats column
car_sales_missing.drop("Seats", axis=1, inplace=True)
car_sales_missing
```

```python
# Shuffle the car_sales_missing with frac = 1.  Save the results to a new variable
shuffled_car_sales_missing = car_sales_missing.sample(frac=1)
shuffled_car_sales_missing
```

```python
# Change the Odometer values from the current kilometers to miles using the lambda function
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].apply(lambda x:x / 1.609)
```

```python
# Now change the title of the Odometer column to read Odometer (miles)
car_sales_missing["Odometer"].rename(columns={"Odometer": "Odometer (miles)"}, inplace=True)
car_sales_missing
```
